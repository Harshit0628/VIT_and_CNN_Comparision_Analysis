```python
# %% [markdown]
# # CNN vs Vision Transformer Comparison
# 
# This notebook performs a comparative analysis between:
# - Pretrained ResNet-50 from torchvision
# - Pretrained ViT-B/16 from timm
# 
# Features:
# - Multi-GPU support (DataParallel)
# - Mixed Precision Training (AMP)
# - Data augmentation
# - Learning rate scheduler and early stopping
# - Comprehensive evaluation metrics
# - Enhanced visualizations

# %% [code]
# Install required packages if they're not already installed
!pip install -q timm

# %% [code]
# Import necessary libraries
import os
import time
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.notebook import tqdm
from datetime import datetime

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import DataLoader
from torchvision import datasets, transforms, models, utils  # Added utils here
from torch.amp import GradScaler, autocast

import timm
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, roc_curve, auc, classification_report,
    precision_recall_curve, average_precision_score
)

# Check if CUDA is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Check for multiple GPUs
num_gpus = torch.cuda.device_count()
print(f"Number of available GPUs: {num_gpus}")

# %% [code]
# Set paths to the dataset
DATA_ROOT = "/kaggle/working/human_dataset/human-and-non-human"
TRAIN_DIR = os.path.join(DATA_ROOT, "training_set/training_set")
TEST_DIR = os.path.join(DATA_ROOT, "test_set/test_set")

# Hyperparameters
BATCH_SIZE = 32
NUM_EPOCHS = 15
LEARNING_RATE = 1e-4
WEIGHT_DECAY = 1e-4
EARLY_STOPPING_PATIENCE = 5
IMAGE_SIZE = 224  # ResNet and ViT-B/16 both accept 224×224 images
NUM_CLASSES = 2  # Binary classification: humans vs non-humans

# %% [code]
# Data Augmentation and Transforms
train_transform = transforms.Compose([
    transforms.Resize((IMAGE_SIZE + 32, IMAGE_SIZE + 32)),  # Resize larger for cropping
    transforms.RandomCrop((IMAGE_SIZE, IMAGE_SIZE)),        # Random crop
    transforms.RandomHorizontalFlip(p=0.5),                # Random flip
    transforms.RandomRotation(15),                         # Random rotation
    transforms.ColorJitter(brightness=0.1, contrast=0.1),  # Color jitter
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],  # ImageNet means
        std=[0.229, 0.224, 0.225]    # ImageNet stds
    )
])

# Validation and Test transforms (no augmentation)
val_transform = transforms.Compose([
    transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

# %% [code]
# Function to visualize a sample batch
def sample_batch_visualization(dataloader, class_names):
    # Get a batch of training data
    inputs, classes = next(iter(dataloader))
    
    # Make a grid from batch
    out = utils.make_grid(inputs[:16])  # Using utils directly since we imported it correctly now
    
    # Convert tensor to numpy array for visualization
    inp = out.numpy().transpose((1, 2, 0))
    
    # Denormalize
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    inp = std * inp + mean
    inp = np.clip(inp, 0, 1)
    
    # Plot
    plt.figure(figsize=(12, 6))
    plt.imshow(inp)
    plt.title("Sample Batch")
    
    # Add class labels
    labels = [class_names[i] for i in classes[:16]]
    plt.figtext(0.5, 0.01, f"Labels: {', '.join(labels)}", ha="center", fontsize=10)
    
    plt.axis('off')
    plt.show()

# %% [code]
# Load datasets
def load_data():
    print("Loading datasets...")
    
    # Load training data
    train_dataset = datasets.ImageFolder(
        root=TRAIN_DIR,
        transform=train_transform
    )
    
    # Calculate split for validation (80% train, 20% validation)
    train_size = int(0.8 * len(train_dataset))
    val_size = len(train_dataset) - train_size
    
    # Split the training dataset
    train_subset, val_subset = torch.utils.data.random_split(
        train_dataset, [train_size, val_size], 
        generator=torch.Generator().manual_seed(42)  # For reproducibility
    )
    
    # Apply validation transform to validation subset
    val_subset.dataset.transform = val_transform
    
    # Load test data
    test_dataset = datasets.ImageFolder(
        root=TEST_DIR,
        transform=val_transform
    )
    
    # Data loaders
    train_loader = DataLoader(
        train_subset, batch_size=BATCH_SIZE, shuffle=True, num_workers=4, pin_memory=True
    )
    val_loader = DataLoader(
        val_subset, batch_size=BATCH_SIZE, shuffle=False, num_workers=4, pin_memory=True
    )
    test_loader = DataLoader(
        test_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=4, pin_memory=True
    )
    
    print(f"Train set size: {len(train_subset)}")
    print(f"Validation set size: {len(val_subset)}")
    print(f"Test set size: {len(test_dataset)}")
    
    class_names = train_dataset.classes
    print(f"Class names: {class_names}")
    
    # Sample batch visualization
    sample_batch_visualization(train_loader, class_names)
    
    return train_loader, val_loader, test_loader, class_names

# %% [code]
# Define models
class ResNetModel(nn.Module):
    def __init__(self, num_classes=2):
        super(ResNetModel, self).__init__()
        # Load pretrained ResNet-50 (using weights parameter instead of deprecated pretrained)
        self.backbone = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V1)
        
        # Replace the final fully connected layer for binary classification
        in_features = self.backbone.fc.in_features
        self.backbone.fc = nn.Sequential(
            nn.Dropout(0.3),  # Add dropout for regularization
            nn.Linear(in_features, num_classes)
        )
    
    def forward(self, x):
        return self.backbone(x)

class ViTModel(nn.Module):
    def __init__(self, num_classes=2):
        super(ViTModel, self).__init__()
        # Load pretrained ViT-B/16 model from timm
        self.backbone = timm.create_model('vit_base_patch16_224', pretrained=True)
        
        # Replace the head for binary classification
        in_features = self.backbone.head.in_features
        self.backbone.head = nn.Sequential(
            nn.Dropout(0.3),  # Add dropout for regularization
            nn.Linear(in_features, num_classes)
        )
    
    def forward(self, x):
        return self.backbone(x)

# %% [code]
# Training and validation functions
def train_epoch(model, dataloader, criterion, optimizer, scaler, epoch):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    
    progress_bar = tqdm(dataloader, desc=f'Epoch {epoch+1} [Train]')
    
    for inputs, labels in progress_bar:
        inputs, labels = inputs.to(device), labels.to(device)
        
        # Zero the parameter gradients
        optimizer.zero_grad()
        
        # Mixed precision training
        with autocast('cuda'):
            outputs = model(inputs)
            loss = criterion(outputs, labels)
        
        # Backward and optimize
        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()
        
        # Statistics
        running_loss += loss.item() * inputs.size(0)
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
        
        # Update progress bar
        progress_bar.set_postfix({'loss': loss.item(), 'acc': 100. * correct / total})
    
    epoch_loss = running_loss / total
    epoch_acc = 100. * correct / total
    
    return epoch_loss, epoch_acc

def validate(model, dataloader, criterion):
    model.eval()
    running_loss = 0.0
    all_preds = []
    all_labels = []
    
    with torch.no_grad():
        for inputs, labels in tqdm(dataloader, desc='Validation'):
            inputs, labels = inputs.to(device), labels.to(device)
            
            # Forward pass
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            
            # Statistics
            running_loss += loss.item() * inputs.size(0)
            _, predicted = torch.max(outputs, 1)
            
            # Store predictions and labels for metrics calculation
            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    
    # Calculate metrics
    val_loss = running_loss / len(dataloader.dataset)
    val_acc = 100. * accuracy_score(all_labels, all_preds)
    
    return val_loss, val_acc, all_preds, all_labels

# %% [code]
# Main training loop with early stopping
def train_model(model, train_loader, val_loader, model_name):
    print(f"\nTraining {model_name}...")
    
    # Initialize criterion, optimizer and scheduler
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.AdamW(model.parameters(), lr=LEARNING_RATE, weight_decay=WEIGHT_DECAY)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=2)
    
    # Initialize mixed precision training
    scaler = GradScaler('cuda')
    
    # Initialize variables for early stopping
    best_val_loss = float('inf')
    best_model_state = None
    early_stopping_counter = 0
    
    # Training metrics history
    train_losses, val_losses = [], []
    train_accs, val_accs = [], []
    learning_rates = []
    start_time = time.time()
    epochs_run = 0
    
    # Training loop
    for epoch in range(NUM_EPOCHS):
        epochs_run = epoch + 1
        epoch_start = time.time()
        
        # Train one epoch
        train_loss, train_acc = train_epoch(model, train_loader, criterion, optimizer, scaler, epoch)
        
        # Validate
        val_loss, val_acc, _, _ = validate(model, val_loader, criterion)
        
        # Learning rate scheduler step
        scheduler.step(val_loss)
        current_lr = optimizer.param_groups[0]['lr']
        learning_rates.append(current_lr)
        
        # Record metrics
        train_losses.append(train_loss)
        val_losses.append(val_loss)
        train_accs.append(train_acc)
        val_accs.append(val_acc)
        
        # Print epoch summary
        epoch_time = time.time() - epoch_start
        print(f"Epoch {epoch+1}/{NUM_EPOCHS} - Time: {epoch_time:.2f}s - "
              f"Train Loss: {train_loss:.4f} - Train Acc: {train_acc:.2f}% - "
              f"Val Loss: {val_loss:.4f} - Val Acc: {val_acc:.2f}% - "
              f"LR: {current_lr:.6f}")
        
        # Check for improvement
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            best_model_state = model.state_dict().copy()
            early_stopping_counter = 0
            print(f"Validation improved! Saving model checkpoint.")
        else:
            early_stopping_counter += 1
            print(f"Validation didn't improve. Early stopping counter: {early_stopping_counter}/{EARLY_STOPPING_PATIENCE}")
            
            if early_stopping_counter >= EARLY_STOPPING_PATIENCE:
                print(f"Early stopping triggered after {epoch+1} epochs")
                break
    
    # Calculate total training time
    total_time = time.time() - start_time
    
    # Load best model state
    model.load_state_dict(best_model_state)
    
    # Return history for plotting and the trained model
    history = {
        'train_loss': train_losses,
        'val_loss': val_losses,
        'train_acc': train_accs,
        'val_acc': val_accs,
        'learning_rates': learning_rates,
        'epochs_run': epochs_run,
        'training_time': total_time,
        'time_per_epoch': total_time / epochs_run
    }
    
    print(f"Training completed in {total_time:.2f} seconds ({history['time_per_epoch']:.2f} s/epoch)")
    
    # Plot learning rate curve
    plt.figure(figsize=(10, 4))
    plt.plot(range(1, len(learning_rates) + 1), learning_rates, 'o-')
    plt.title(f'{model_name} Learning Rate Schedule')
    plt.xlabel('Epoch')
    plt.ylabel('Learning Rate')
    plt.grid(True, alpha=0.3)
    plt.show()
    
    return model, history

# %% [code]
# Evaluation function
def evaluate_model(model, test_loader, class_names):
    model.eval()
    all_preds = []
    all_labels = []
    all_probs = []  # For ROC curve
    
    with torch.no_grad():
        for inputs, labels in tqdm(test_loader, desc='Evaluating'):
            inputs, labels = inputs.to(device), labels.to(device)
            
            # Forward pass
            outputs = model(inputs)
            probs = F.softmax(outputs, dim=1)
            _, predicted = torch.max(outputs, 1)
            
            # Store predictions, probabilities and labels
            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
            all_probs.extend(probs[:, 1].cpu().numpy())  # Probability of positive class
    
    # Calculate metrics
    accuracy = accuracy_score(all_labels, all_preds)
    precision = precision_score(all_labels, all_preds, average='binary')
    recall = recall_score(all_labels, all_preds, average='binary')
    f1 = f1_score(all_labels, all_preds, average='binary')
    
    # Calculate confusion matrix
    cm = confusion_matrix(all_labels, all_preds)
    
    # Calculate ROC curve and AUC
    fpr, tpr, _ = roc_curve(all_labels, all_probs)
    roc_auc = auc(fpr, tpr)
    
    # Calculate precision-recall curve and average precision
    precision_curve, recall_curve, _ = precision_recall_curve(all_labels, all_probs)
    avg_precision = average_precision_score(all_labels, all_probs)
    
    # Print classification report
    print("\nClassification Report:")
    print(classification_report(all_labels, all_preds, target_names=class_names))
    
    # Return metrics
    metrics = {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1_score': f1,
        'confusion_matrix': cm,
        'roc': {'fpr': fpr, 'tpr': tpr, 'auc': roc_auc},
        'pr_curve': {'precision': precision_curve, 'recall': recall_curve, 'ap': avg_precision},
        'predictions': all_preds,
        'labels': all_labels,
        'probabilities': all_probs
    }
    
    return metrics

# %% [code]
# Visualization functions
def plot_training_history(cnn_history, vit_history):
    fig, axs = plt.subplots(2, 2, figsize=(15, 10))
    
    # Determine max epochs to plot
    max_epochs = max(cnn_history['epochs_run'], vit_history['epochs_run'])
    
    # Plot Training Loss
    axs[0, 0].plot(cnn_history['train_loss'], 'o-', label='CNN')
    axs[0, 0].plot(vit_history['train_loss'], 'o-', label='ViT')
    axs[0, 0].set_title('Training Loss')
    axs[0, 0].set_xlabel('Epoch')
    axs[0, 0].set_ylabel('Loss')
    axs[0, 0].legend()
    axs[0, 0].set_xlim(0, max_epochs-1)
    axs[0, 0].grid(alpha=0.3)
    
    # Plot Validation Loss
    axs[0, 1].plot(cnn_history['val_loss'], 'o-', label='CNN')
    axs[0, 1].plot(vit_history['val_loss'], 'o-', label='ViT')
    axs[0, 1].set_title('Validation Loss')
    axs[0, 1].set_xlabel('Epoch')
    axs[0, 1].set_ylabel('Loss')
    axs[0, 1].legend()
    axs[0, 1].set_xlim(0, max_epochs-1)
    axs[0, 1].grid(alpha=0.3)
    
    # Plot Training Accuracy
    axs[1, 0].plot(cnn_history['train_acc'], 'o-', label='CNN')
    axs[1, 0].plot(vit_history['train_acc'], 'o-', label='ViT')
    axs[1, 0].set_title('Training Accuracy')
    axs[1, 0].set_xlabel('Epoch')
    axs[1, 0].set_ylabel('Accuracy (%)')
    axs[1, 0].legend()
    axs[1, 0].set_xlim(0, max_epochs-1)
    axs[1, 0].grid(alpha=0.3)
    
    # Plot Validation Accuracy
    axs[1, 1].plot(cnn_history['val_acc'], 'o-', label='CNN')
    axs[1, 1].plot(vit_history['val_acc'], 'o-', label='ViT')
    axs[1, 1].set_title('Validation Accuracy')
    axs[1, 1].set_xlabel('Epoch')
    axs[1, 1].set_ylabel('Accuracy (%)')
    axs[1, 1].legend()
    axs[1, 1].set_xlim(0, max_epochs-1)
    axs[1, 1].grid(alpha=0.3)
    
    plt.tight_layout()
    plt.show()

def plot_confusion_matrices(cnn_cm, vit_cm, class_names):
    fig, axs = plt.subplots(1, 2, figsize=(12, 5))
    
    # Plot CNN confusion matrix
    sns.heatmap(cnn_cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, 
                yticklabels=class_names, ax=axs[0])
    axs[0].set_title('CNN Confusion Matrix')
    axs[0].set_ylabel('True Label')
    axs[0].set_xlabel('Predicted Label')
    
    # Plot ViT confusion matrix
    sns.heatmap(vit_cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, 
                yticklabels=class_names, ax=axs[1])
    axs[1].set_title('ViT Confusion Matrix')
    axs[1].set_ylabel('True Label')
    axs[1].set_xlabel('Predicted Label')
    
    plt.tight_layout()
    plt.show()

def plot_roc_curves(cnn_roc, vit_roc):
    plt.figure(figsize=(10, 8))
    
    # Plot CNN ROC curve
    plt.plot(cnn_roc['fpr'], cnn_roc['tpr'], color='blue', linewidth=2,
             label=f'CNN (AUC = {cnn_roc["auc"]:.3f})')
    
    # Plot ViT ROC curve
    plt.plot(vit_roc['fpr'], vit_roc['tpr'], color='green', linewidth=2,
             label=f'ViT (AUC = {vit_roc["auc"]:.3f})')
    
    # Plot diagonal
    plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
    
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve Comparison')
    plt.legend(loc="lower right")
    plt.grid(True, alpha=0.3)
    plt.show()

def plot_precision_recall_curves(cnn_pr, vit_pr):
    plt.figure(figsize=(10, 8))
    
    # Plot CNN PR curve
    plt.plot(cnn_pr['recall'], cnn_pr['precision'], color='blue', linewidth=2,
             label=f'CNN (AP = {cnn_pr["ap"]:.3f})')
    
    # Plot ViT PR curve
    plt.plot(vit_pr['recall'], vit_pr['precision'], color='green', linewidth=2,
             label=f'ViT (AP = {vit_pr["ap"]:.3f})')
    
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Precision-Recall Curve Comparison')
    plt.legend(loc="best")
    plt.grid(True, alpha=0.3)
    plt.show()

def display_comparison_table(cnn_metrics, vit_metrics, cnn_history, vit_history):
    # Create comparison table
    table_data = [
        ["Model", "Test Accuracy", "F1 Score", "Precision", "Recall", "Training Time", "ROC-AUC", "Avg Precision"],
        ["ResNet50", f"{cnn_metrics['accuracy']:.4f}", f"{cnn_metrics['f1_score']:.4f}", 
         f"{cnn_metrics['precision']:.4f}", f"{cnn_metrics['recall']:.4f}", 
         f"{cnn_history['training_time']:.2f}s", f"{cnn_metrics['roc']['auc']:.4f}", 
         f"{cnn_metrics['pr_curve']['ap']:.4f}"],
        ["ViT-B/16", f"{vit_metrics['accuracy']:.4f}", f"{vit_metrics['f1_score']:.4f}", 
         f"{vit_metrics['precision']:.4f}", f"{vit_metrics['recall']:.4f}", 
         f"{vit_history['training_time']:.2f}s", f"{vit_metrics['roc']['auc']:.4f}",
         f"{vit_metrics['pr_curve']['ap']:.4f}"]
    ]
    
    # Print table
    row_format = "{:<10} {:<15} {:<15} {:<15} {:<15} {:<15} {:<15} {:<15}"
    print("\n" + "="*120)
    print("MODEL COMPARISON")
    print("="*120)
    for row in table_data:
        print(row_format.format(*row))
    print("="*120)
    
    # Create a visual bar chart comparison
    metrics_names = ["Accuracy", "F1 Score", "Precision", "Recall", "ROC-AUC", "Avg Precision"]
    cnn_values = [
        cnn_metrics['accuracy'], 
        cnn_metrics['f1_score'], 
        cnn_metrics['precision'], 
        cnn_metrics['recall'], 
        cnn_metrics['roc']['auc'],
        cnn_metrics['pr_curve']['ap']
    ]
    
    vit_values = [
        vit_metrics['accuracy'], 
        vit_metrics['f1_score'], 
        vit_metrics['precision'], 
        vit_metrics['recall'], 
        vit_metrics['roc']['auc'],
        vit_metrics['pr_curve']['ap']
    ]
    
    # Plot bar chart
    fig, ax = plt.subplots(figsize=(12, 6))
    x = np.arange(len(metrics_names))
    width = 0.35
    
    rects1 = ax.bar(x - width/2, cnn_values, width, label='CNN (ResNet-50)')
    rects2 = ax.bar(x + width/2, vit_values, width, label='ViT-B/16')
    
    ax.set_title('Model Performance Metrics Comparison')
    ax.set_ylabel('Score')
    ax.set_xticks(x)
    ax.set_xticklabels(metrics_names)
    ax.legend()
    ax.grid(axis='y', alpha=0.3)
    
    # Add value labels on top of bars
    def add_labels(rects):
        for rect in rects:
            height = rect.get_height()
            ax.annotate(f'{height:.3f}',
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 3),  # 3 points vertical offset
                        textcoords="offset points",
                        ha='center', va='bottom', rotation=90)
    
    add_labels(rects1)
    add_labels(rects2)
    
    plt.tight_layout()
    plt.show()
    
    # Also save as CSV for further analysis
    with open('model_comparison.csv', 'w') as f:
        for row in table_data:
            f.write(','.join(str(item) for item in row) + '\n')

# %% [code]
# New function to visualize model prediction samples
def visualize_predictions(model, test_loader, class_names, model_name, num_images=8):
    model.eval()
    images_so_far = 0
    plt.figure(figsize=(12, 8))
    
    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs = inputs.to(device)
            labels = labels.to(device)
            
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)
            
            for j in range(inputs.size()[0]):
                if images_so_far >= num_images:
                    break
                    
                images_so_far += 1
                ax = plt.subplot(num_images//4, 4, images_so_far)
                ax.axis('off')
                
                # Denormalize image
                img = inputs.cpu().data[j].numpy().transpose((1, 2, 0))
                mean = np.array([0.485, 0.456, 0.406])
                std = np.array([0.229, 0.224, 0.225])
                img = std * img + mean
                img = np.clip(img, 0, 1)
                
                # Display image
                ax.imshow(img)
                
                # Add prediction vs ground truth
                pred_label = class_names[preds[j]]
                true_label = class_names[labels[j]]
                
                # Set color based on correctness
                color = 'green' if pred_label == true_label else 'red'
                ax.set_title(f'P: {pred_label}\nT: {true_label}', color=color)
                
            if images_so_far >= num_images:
                break
                
    plt.suptitle(f'{model_name} Predictions on Test Data', fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust layout to make room for suptitle
    plt.show()

# %% [code]
# Function to visualize feature space using t-SNE
# Function to visualize feature space using t-SNE
def visualize_feature_space(model, dataloader, class_names, model_name, num_samples=500):
    from sklearn.manifold import TSNE
    import torch
    import torch.nn as nn
    import numpy as np
    import matplotlib.pyplot as plt
    
    # Function to extract features from the penultimate layer
    def extract_features(model, dataloader, num_samples):
        model.eval()
        features = []
        labels = []
        samples_processed = 0
        
        if isinstance(model, nn.DataParallel):
            base_model = model.module
        else:
            base_model = model
            
        with torch.no_grad():
            for inputs, targets in dataloader:
                inputs, targets = inputs.to(device), targets.to(device)
                
                if isinstance(base_model, ResNetModel):
                    feature_extractor = nn.Sequential(*list(base_model.backbone.children())[:-1])
                    batch_features = feature_extractor(inputs)
                    # Flatten features
                    batch_features = batch_features.reshape(batch_features.size(0), -1)
                # For ViT
                else:
                    # Get features before the classification head
                    batch_features = base_model.backbone.forward_features(inputs)
                    # Extract class token (CLS) - usually the first token for ViTs
                    if batch_features.dim() > 2:  # If it's a 3D tensor [batch, tokens, features]
                        batch_features = batch_features[:, 0]  # Get only the CLS token
                        
                # Store features and labels
                features.append(batch_features.cpu())
                labels.append(targets.cpu())
                
                # Count samples
                samples_processed += inputs.size(0)
                if samples_processed >= num_samples:
                    break
        
        # Concatenate all features and labels
        features = torch.cat(features, dim=0).numpy()
        labels = torch.cat(labels, dim=0).numpy()
        
        # Ensure features are 2D for t-SNE
        if features.ndim > 2:
            features = features.reshape(features.shape[0], -1)
            
        return features[:num_samples], labels[:num_samples]
    
    # Extract features
    print(f"Extracting features for {model_name}...")
    features, labels = extract_features(model, dataloader, num_samples)
    
    # Apply t-SNE for dimensionality reduction
    print("Applying t-SNE dimensionality reduction...")
    tsne = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=1000)
    features_embedded = tsne.fit_transform(features)
    
    # Plot the embedding
    plt.figure(figsize=(10, 8))
    for i, label in enumerate(np.unique(labels)):
        idx = labels == label
        plt.scatter(features_embedded[idx, 0], features_embedded[idx, 1], 
                    label=class_names[i], alpha=0.7)
    
    plt.title(f"{model_name} Feature Space Visualization (t-SNE)")
    plt.legend()
    plt.grid(alpha=0.3)
    plt.show()
    
# %% [code]
# Main execution
def main():
    # Set random seed for reproducibility
    torch.manual_seed(42)
    np.random.seed(42)
    
    # Load data
    train_loader, val_loader, test_loader, class_names = load_data()
    
    # Create models
    print("\nInitializing models...")
    
    # Initialize ResNet model
    cnn_model = ResNetModel(num_classes=NUM_CLASSES)
    
    # Use DataParallel if multiple GPUs are available
    if num_gpus > 1:
        print(f"Using {num_gpus} GPUs for CNN training...")
        cnn_model = nn.DataParallel(cnn_model)
    
    cnn_model = cnn_model.to(device)
    
    # Initialize ViT model
    vit_model = ViTModel(num_classes=NUM_CLASSES)
    
    # Use DataParallel if multiple GPUs are available
    if num_gpus > 1:
        print(f"Using {num_gpus} GPUs for ViT training...")
        vit_model = nn.DataParallel(vit_model)
    
    vit_model = vit_model.to(device)
    
    # Train models
    cnn_model, cnn_history = train_model(cnn_model, train_loader, val_loader, "ResNet-50")
    vit_model, vit_history = train_model(vit_model, train_loader, val_loader, "ViT-B/16")
    
    # Evaluate models
    print("\nEvaluating ResNet-50...")
    cnn_metrics = evaluate_model(cnn_model, test_loader, class_names)
    
    print("\nEvaluating ViT-B/16...")
    vit_metrics = evaluate_model(vit_model, test_loader, class_names)
    
    # Visualize results
    print("\nGenerating comparison visualizations...")
    plot_training_history(cnn_history, vit_history)
    plot_confusion_matrices(cnn_metrics['confusion_matrix'], vit_metrics['confusion_matrix'], class_names)
    plot_roc_curves(cnn_metrics['roc'], vit_metrics['roc'])
    plot_precision_recall_curves(cnn_metrics['pr_curve'], vit_metrics['pr_curve'])
    display_comparison_table(cnn_metrics, vit_metrics, cnn_history, vit_history)
    
    # Visualize model predictions
    visualize_predictions(cnn_model, test_loader, class_names, "ResNet-50")
    visualize_predictions(vit_model, test_loader, class_names, "ViT-B/16")
    
    # Visualize feature space
    visualize_feature_space(cnn_model, test_loader, class_names, "ResNet-50")
    visualize_feature_space(vit_model, test_loader, class_names, "ViT-B/16")
    
    # Print conclusion
    print("\n" + "="*120)
    print("CONCLUSION")
    print("="*120)
    
    # Determine which model performed better
    if vit_metrics['accuracy'] > cnn_metrics['accuracy']:
        winner = "ViT-B/16"
        accuracy_diff = (vit_metrics['accuracy'] - cnn_metrics['accuracy']) * 100
    else:
        winner = "ResNet-50"
        accuracy_diff = (cnn_metrics['accuracy'] - vit_metrics['accuracy']) * 100
    
    print(f"The {winner} model performed better with a {accuracy_diff:.2f}% higher accuracy.")
    print(f"Key metrics for the winning model ({winner}):")
    
    if winner == "ViT-B/16":
        print(f"  - Accuracy: {vit_metrics['accuracy']:.4f}")
        print(f"  - F1 Score: {vit_metrics['f1_score']:.4f}")
        print(f"  - Training time: {vit_history['training_time']:.2f} seconds")
    else:
        print(f"  - Accuracy: {cnn_metrics['accuracy']:.4f}")
        print(f"  - F1 Score: {cnn_metrics['f1_score']:.4f}")
        print(f"  - Training time: {cnn_history['training_time']:.2f} seconds")
    
    # Save models
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    torch.save(cnn_model.state_dict(), f"resnet50_model_{timestamp}.pth")
    torch.save(vit_model.state_dict(), f"vit_model_{timestamp}.pth")
    print(f"\nModels saved with timestamp: {timestamp}")
    
    print("="*120)

# %% [code]
# Execute main function
if __name__ == "__main__":
    print(f"Starting CNN vs Vision Transformer comparison...")
    start_time = time.time()
    
    main()
    
    total_time = time.time() - start_time
    print(f"\nTotal execution time: {total_time:.2f} seconds ({total_time/60:.2f} minutes)")
                
```
